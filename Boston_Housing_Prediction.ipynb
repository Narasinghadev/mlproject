```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn.datasets
from sklearn.model_selection import train_test_split
from xgboost import XGBRegressor
from sklearn import metrics
```

```python
# Import the dataset

house_price_dataset = pd.read_csv('/content/BostonHousing.csv')
```

```python
print(house_price_dataset)
```

```python
# Loading the dataset to a Pandas Dataframe
house_price_dataset.head()
```

```python
# Check the number of rows and columsn
house_price_dataset.shape

```

```python
# Checking for the missing values

house_price_dataset.isnull().sum()
```

```python
#  Statistical measures of the dataframe
house_price_dataset.describe()
```

Understanding the correlation of the features
1. Positive Correleration
2. Negative Correleration

```python
correleration = house_price_dataset.corr()
```

```python
 # Constructing a heatmap  to understand the correleration
 plt.figure(figsize=(10,10))
 sns.heatmap(correleration, cbar= True, square= True , fmt= '.1f', annot = True, annot_kws={'size':8}, cmap = 'Blues')
```

```python
 # Splitting the Data and the Target

 X = house_price_dataset.drop(['medv'] , axis = 1 )
 Y= house_price_dataset['medv']
```

```python
print(X)
print(Y)
```

Splitting the Data into the Training Data and the Test Data

```python
X_train, X_test , Y_train , Y_test = train_test_split(X, Y , test_size =0.2 , random_state= 2 )
```

```python
print (X_train.shape)
```

Train the Model , XGBoost Regressor , Model Training

```python
 # Loading the Model

 model = XGBRegressor()
```

```python
model.fit(X_train, Y_train)
```

Prediction on the Training Data

```python
 # Accuracy for the Prediction on the Training Data

 training_data = model.predict(X_train)
```

```python
print(training_data)
```

```python
 # R sqauare Model and Mean sqaure are used for the Regression Model

 score_1 = metrics.r2_score(Y_train , training_data)


  # Mean Absolute Error

score_2 = metrics.mean_absolute_error(Y_train , training_data)
```

```python
print ('R square Error :' ,score_1)
print (' Mean Absolute Error :', score_2)
```

Visulaizing the Actual Prices and the Predicted Prices

```python
plt.scatter(Y_train, training_data)
plt.xlabel('Actual Prices')
plt.ylabel('Predicted Prices')
plt.title('Actual vs Predicted Price')
plt.show()
```

Evaluation on the Test Data

```python
 test_data = model.predict(X_test)
```

Find the Error in the Test Value ,  # Test Data Check Value

```python
 score_4 = metrics.r2_score(Y_test , test_data)


  # Mean Absolute Error

score_5 = metrics.mean_absolute_error(Y_test , test_data)
```

```python
print ('R square Error :' ,score_4)
print (' Mean Absolute Error :', score_5)
```
